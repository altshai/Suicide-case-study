```python
df = df[~df['State'].isin(['Total (All India)', 'Total (States)', 'Total (Uts)'])]
print(df.head())
```

```python
print(df['State'].unique())
```

```python
statewise_df = df.drop(df[(df.State == 'Total (Uts)') | (df.State == 'Total (All India)') | 
               (df.State == 'Total (States)')].index)
```

```python
statewise_df.replace('Bankruptcy or Sudden change in Economic', 
           'Bankruptcy or Sudden change in Economic Status', inplace=True)
statewise_df.replace('By Other means (please specify)', 'By Other means', inplace=True)
statewise_df.replace('Not having Children(Barrenness/Impotency',
           'Not having Children (Barrenness/Impotency)', inplace=True)
```

```python
statewise_df = statewise_df.reset_index(drop=True)
statewise_df
```

```python
import seaborn as sns  #https://seaborn.pydata.org/generated/seaborn.set_style.html
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
#We use rc i.e a runtime configuration (rc) containing the default styles for every plot element you create
matplotlib.rcParams['font.size'] = 12
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

Q.1 Yearwise Total Number Of Deaths

```python
yearwise_total_suicide = statewise_df.groupby("Year")["Total"].sum()
print('Total number of suicides in each year: ', yearwise_total_suicide)
```

```python
# Plotting the yearwise total suicides in a line graph
yearwise_total_suicide.plot(kind="line", figsize=(9, 5), color='r')
plt.title('Yearwise Total Suicides')
plt.xlabel('Year')
plt.ylabel('Number of suicides')
plt.show()
```

```python
year_with_highest_suicides = yearwise_total_suicide.idxmax()
highest_suicides = yearwise_total_suicide.max()
print(f"The highest number of suicides occurred in the year {year_with_highest_suicides} with {highest_suicides} suicides.")
```

The highest number of suicides occurred in the year 2011 with 677159 suicides.

Q2. Statewise Number Of Suicides

```python
statewise_total_suicide = statewise_df.groupby("State")["Total"].sum()
statewise_total_suicide
```

```python
# Plotting the statewise total suicides in a bar graph
statewise_total_suicide.plot(kind="bar", figsize=(15, 7), color='b')
plt.title('Statewise Total Suicides')
plt.xlabel('State')
plt.ylabel('Number of suicides')

# Adding the total count on each bar
for index, value in enumerate(statewise_total_suicide):
    plt.text(index, value, str(value), ha='center', va='bottom', fontsize=8)
plt.xticks(rotation=90)
plt.show()
```

maharshtra state is the on top for suicide with 901945 suicides

Q3 Suicides According To The Causes In Each Year

```python
# Grouping the data by Year and Type and summing the Total suicides
causes_yearwise = statewise_df.groupby(['Year', 'Type'])['Total'].sum().unstack()

# Plotting the data
causes_yearwise.plot(kind='bar', stacked=True, figsize=(15, 7))
plt.title('Suicides According To The Causes In Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Suicides')
plt.legend(title='Cause', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
causewise_total_suicide = statewise_df.groupby("Type")["Total"].sum()
print(causewise_total_suicide)
```

```python
# Plotting the causewise total suicides in a bar graph
causewise_total_suicide.plot(kind="barh", figsize=(30, 30), color='g')
plt.title('Causewise Total Suicides')
plt.xlabel('Number of Suicides')
plt.ylabel('Cause')

# Adding the total count on each bar
for index, value in enumerate(causewise_total_suicide):
    plt.text(value, index, str(value), ha='left', va='center', fontsize=8)
plt.show()
```

Q.4 Suicide According To Gender

```python
genderwise_suicide = statewise_df.groupby("Gender")["Total"].sum()
print(genderwise_suicide)
```
